O `Turtle` é um módulo do Python que permite criar imagens e formas em uma tela virtual.

A tartaruga pode se mover em todas as direções no plano $Oxy$. Os deslocamentos horizontais e verticais são programáveis ao se determinar:

* movimentos para frente ou para trás, com distâncias definidas.

* rotações para a esquerda ou para a direita, medidas em graus.

Para criar uma tartaruga chamda `donatelo` (na verdade um objeto do tipo `Turtle`, referenciado pela variável `donatelo`), usamos a sintaxe:

`donatelo = Turtle()`

`donatelo` nasce sabendo realizar determinadas ações (movimentar-se ou girar). O programador invoca as ações desejadas, ativando os métodos definidos para o objeto `donatelo`. As propriedades (ou atributos) de `donatelo` podem ser ajustadas (tais como forma, cores, espessura da caneta, direção, etc).

Execute a célula a seguir e veja `donatelo` em ação.

```python
# Constantes
LADO = 150            # Lado da "janela" (canvas)
VELOCIDADE = 2        # Escala de 1 a 13

# Criação do canvas
clearscreen()             # Limpa a tela da execução anterior
setup(2*LADO, 2*LADO)     # Tamanho da janela
bgcolor('lightgray')      # Cor de fundo da tela
showborder()              # Mostra a borda

# Criação da tartaruga donatelo
donatelo = Turtle()         # Criação
donatelo.speed(VELOCIDADE)  # Alterando a velocidade

# Movimentação
donatelo.forward(100)       # donatelo.fd(100)
donatelo.left(45)           # donatelo.lt(45)
donatelo.forward(20)
donatelo.right(65)          # donatelo.rt(65)
donatelo.backward(150)      # donatelo.bk(150)
```

Também é possível escrever usando o módulo `Turtle`. Examine o código a seguir e o execute. Pesquise sobre o método `write`.

```python
# Constantes
LADO = 300        # Lado da "janela" (canvas)
VELOCIDADE = 7    # Escala de 1 a 13

# Criação do canvas
clearscreen()             # Limpa a tela da execução anterior
setup(LADO, LADO)         # Tamanho da janela
bgcolor('lightblue')      # Cor de fundo da tela
showborder()              # Mostra a borda

# Criação da tartaruga
jabu = Turtle()           # Criação do jabu
jabu.speed(VELOCIDADE)    # Alterando a velocidade do jabu
jabu.hideturtle()         # Esconde o jabu

# Posicionando para escrita
jabu.penup()              # Levanta a caneta, permitindo o movimento sem traços
jabu.goto(0, LADO/6)      # Posiciona jabu na tela (a origem é o centro da tela)

# Primeira palavra
fonte1 = (60, "Comic Sans", "italic")
jabu.write("Lógica ", align="center", font=fonte1)
# Segunda palavra
jabu.goto(0, 0)
fonte2 = (40, "Courier", "normal")
jabu.write("de", align="center", font=fonte2)
# Terceira palavra
jabu.goto(0, -LADO/6)
fonte3 = (50, "Times New Roman", "bold")
jabu.write("Programação", align="center", font=fonte3)
```
