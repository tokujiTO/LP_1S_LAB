### Exemplos

Exemplo 1: Uma pessoa deseja criar uma poupança de 12350 reais para uma viagem. Todo mês ela consegue economizar 832 reais. Quantos meses ela irá demorar para juntar o dinheiro necessário?

```python
meta = 12350
economia_por_mes = 832
total_acumulado = 0
meses_passados = 0

while total_acumulado < meta:
    total_acumulado += economia_por_mes
    meses_passados += 1

print("Levará", meses_passados, "meses para atingir a meta financeira.")
```

Exemplo 2: Imagine que você está criando um jogo simples em que o computador escolhe um número aleatório entre 1 e 100, e o jogador precisa adivinhar esse número. Cada vez que o jogador faz uma tentativa, o computador fornece uma dica se o número correto é maior ou menor. O jogo continua até que o jogador adivinhe o número correto.

```python
import random

# Gerar um número aleatório entre 1 e 100
numero_correto = random.randint(1, 100)

# Inicializar variáveis
tentativas = 0
adivinhou = False

while not adivinhou:
    # Solicitar uma tentativa ao jogador
    tentativa = int(input("Tente adivinhar o número: "))

    # Atualizar o número de tentativas
    tentativas += 1

    # Verificar se a tentativa está correta
    if tentativa == numero_correto:
        adivinhou = True
    elif tentativa < numero_correto:
        print("Tente um número maior.")
    else:
        print("Tente um número menor.")

print(f"Parabéns! O número correto era {numero_correto}. Você adivinhou corretamente em {tentativas} tentativas.")
```

1 - Validação de Entrada do Usuário

Peça ao usuário para digitar um número dentro de um determinado intervalo (p. ex., entre 1 e 10). Use um while loop para continuar solicitando a entrada até que um valor válido seja fornecido. Em outras palavras, enquanto não for digitado um valor dentro desse intervalo, o programa continua solicitando a entrada do usuário.

```python
num = int(input("Digite um número no intervalo de 1 a 10: "))
while not 0 <= num <= 10:
    num = int(input("DIgite um número no intervalo de 1 a 10: "))
```

3 - Simulação de Crescimento Populacional

Suponha que uma população de organismos cresce exponencialmente a uma certa taxa por período. Use um while loop para simular o crescimento populacional até que ela atinja um limite desejado.

```python
base = int(input("Digite o valor inicial do crescimento populacional: "))
indice = int(input("Digite o expoente q o numero será elevado(ritmo de crescimento): "))
limit = int(input("Digite o limite populacional: "))
while base <= limit:
    base = base**indice
    print(base)
```

4 - Cálculo de Juros Compostos:

Simule o crescimento de um investimento ao longo do tempo com juros compostos. Use um while loop para calcular quanto tempo levará para o investimento dobrar.

Fórmula para o investimento: Montante = Capital * ((1 + taxa)**periodo)

```python
cap = int(input("Digite o capital inicial: R$"))
p = 1
mont = 1
tax = float(input("Qual a taxa? "))
while mont <= 2 * cap:
    mont = cap * ((1+tax)**p)
    p += 1
print(p, mont)
```

5 - Descobrindo o Mínimo Múltiplo Comum (MMC):

Dados dois números, use um while loop para encontrar o Mínimo Múltiplo Comum (MMC) entre eles.

```python
a = int(input("Digite um número: "))
b = int(input("Digite outro número: "))
c = 1
while c%a!=0 or c%b!=0:
    c += 1
print(f"O mmc de {a} e {b} é {c}") 
```

1 - Calculando a Média de uma Lista de Notas:

Dada uma lista de notas de alunos, use um for loop para calcular a média das notas.

```python
limit = int(input("Quantos alunos são? "))
somaNotas = 0
for c in range(1, limit+1):
    somaNotas += int(input(f"Digite a nota do aluno {c}: "))
print(f'A média de notas foi {somaNotas/limit}')
```

2 - Contagem de Caracteres em uma String

Dada uma string, use um for loop para contar quantas vezes um determinado caractere aparece nela.

```python
count = 0
for c in "itaquaquecetuba":
    if c == "a":
        count += 1
print(count)
```

3 - Tabuada de um Número

Dado um número, use um for loop para gerar a tabuada desse número até um determinado limite.

```python
num = int(input("Digite um numero: "))
limit = int(input('Deseja ver a tabuada até que numero? '))
for c in range(1, limit+1):
    print(f"{num} X {c} = {num*c}")
```

4 - Soma dos Números Pares em um Intervalo

Dado um intervalo de números, use um for loop para calcular a soma dos números pares dentro desse intervalo.

```python
somaDosNumeros = 0
limit = int(input("Até onde deseja somar? "))
for c in range(0, limit+1):
    if c%2==0:
        somaDosNumeros += c
print(f"A soma dos números pares é {somaDosNumeros}")
```

5 - Verificando Números Primos em uma Lista

Dada uma lista de números, use um for loop para verificar quais são números primos.

```python
lista = [1, 2, 5, 3, 9, 7, 13]
for c in lista:
    count = 0
    for i in range(2, c):
        if c % i == 0:
            count += 1
    if count == 0:
        print(c)
```

6 - Formatando uma Lista de Nomes

Dada uma lista de nomes, use um for loop para formatar cada nome (por exemplo, capitalizando a primeira letra).

```python
lista = ["tiago", 'pedro', 'guilherme', 'gabriel']
for i in range(len(lista)):
    lista[i] = lista[i].upper()
print(lista)
```

7 - Gerando Sequência de Fibonacci

Use um for loop para gerar os primeiros N termos da sequência de Fibonacci.

```python
limit = int(input("Digite o limite: "))
fibo1 = 1
fibo2 = 1
print(fibo1)
print(fibo2)
for c in range(limit-2):
    fibo3 = fibo1 + fibo2
    fibo2 = fibo1
    fibo1 = fibo3 
    print(fibo3)
```

1.  Crie um programa que procura pelo número 7 na sequência [2, 4, 6, 8, 10, 7, 12, 14, 16] e imprime a posição quando o número é encontrado.

```python
lista = [2, 4, 6, 8, 10, 7, 12, 14, 16]
cont = -1
for c in lista:
    cont+=1
    if c == 7:
        print(f'O numero foi encontrado na posição {cont}')
```

```python
numero = int(input('Digite um número para saber se é primo: '))
for c in range(2, numero):
    if numero % c == 0:
        print('O número não é primo')
        break
```
